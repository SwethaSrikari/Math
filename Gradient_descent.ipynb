A small demonstration showing how gradient descent works using Python

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
ips = np.array(range(-10, 11))

# Quadratic function - x^2
def quadratic(x):
  return (x) ** 2.0

ops = quadratic(ips)

# Derivative of quadratic function (x^2) - 2x
def quad_derivative(input):
  return 2.0 * input
````

````python
plt.plot(ips, ops)
plt.xlabel('inputs')
plt.ylabel('outputs')
plt.title('Quadratic function - x^2')
````

In the context of Machine learning: Gradient descent is an optimization algorithm used to find parameters that minimize the loss/cost/objective function.

How does gradient descent exactly do that? - The algorithm updates the parameters (inputs to the objective function) such that the objective function moves in the negative gradient direction to reach the minimum objective function.

The direction of the gradient affects the way weights are updated in order to find the minimum point. **In case of gradient descent** - If the gradient is positive, weights are reduced to reach the minimum. If the gradient is negative, weights are increased.

The weights can be updated using this formula - 
```
weights = weights - step_size * derivative of the objective function
```

Step size decides how fast/ how big steps you want to take to reach the minimum. The beauty of this formula is that it automatically increases or decreases the weights depending on the direction (positive or negative) of the derivative.

````python
def gradient_descent(wt, loss, step_size):
  """
  wt (float): random weight
  loss (float): objective function; in this case, quadratic(wt) - wt^2
  step_size (float): learning_rate
  """
  wts, losses = [wt], [loss]
  # Keep updating weights until objective function reaches 0 (max 5000 iterations)
  while quadratic(wt) != 0: # End search when objective function reaches 0
    wt = wt - step_size * quad_derivative(wt)
    wts.append(wt)
    losses.append(quadratic(wt))
    if len(wts) > 4000: # Max iterations
      break
  return wts, losses
````

The red dots showing the steps taken to reach the minimum. With low step sizes, more number of steps are needed to reach the minimum. As the step size increases, the number of steps reduce and then increase again after the optimal step size range passes.




````python
plt.plot(ips, ops)
wts, losses = gradient_descent(-10.0, 100.0, 0.1)
plt.plot(wts, losses, '.-', color='red')
plt.xlabel('weight')
plt.ylabel('loss_function')
plt.title(f'Gradient descent with {len(wts)-1} steps')
````

Optimal step size for this case is 0.5

````python
plt.plot(ips, ops)
wts, losses = gradient_descent(10.0, 100.0, 0.5)
plt.plot(wts, losses, '.-', color='red')
plt.xlabel('weight')
plt.ylabel('loss_function')
plt.title(f'Gradient descent with {len(wts)-1} steps')
````

If the step size a.k.a learning rate is high, the loss overshoots and keeps bouncing for a while before reaching the minimum or sometimes just keeps bouncing.

````python
plt.plot(ips, ops)
wts, losses = gradient_descent(10.0, 100.0, 0.8)
plt.plot(wts, losses, '.-', color='red')
plt.xlabel('weight')
plt.ylabel('loss_function')
plt.title(f'Gradient descent with {len(wts)-1} steps')
````

````python
plt.plot(ips, ops)
wts, losses = gradient_descent(10.0, 100.0, 0.95)
plt.plot(wts, losses, '.-', color='red')
plt.xlabel('weight')
plt.ylabel('loss_function')
plt.title(f'Gradient descent with {len(wts)-1} steps')
````

In this case with step size = 1, the weight keeps shifting endlessly between -10 and 10 and the loss never reaches 0.

````python
plt.plot(ips, ops)
wts, losses = gradient_descent(10.0, 100.0, 1)
plt.plot(wts, losses, '.-', color='red')
plt.xlabel('weight')
plt.ylabel('loss_function')
plt.title(f'Gradient descent with {len(wts)-1} steps')
````
